## 2. Loading Data with `pd.read_csv()`

`pd.read_csv()` reads a comma-separated file and returns a **DataFrame** —
think of it as a spreadsheet table inside Python.

The only required argument is the file path.

```python
# Load the dataset — adjust the path if your file is in a different folder
df = pd.read_csv("World_Dev_Indicators.csv")

print("Data loaded successfully!")
print(f"Shape: {df.shape[0]} rows x {df.shape[1]} columns")
```

```python
# Show all column names
print("Columns:")
for col in df.columns:
    print(" -", col)
```

## 3. Key Arguments of `pd.read_csv()`

`pd.read_csv()` has many optional arguments. Here are the most useful ones for beginners.

```python
# nrows — load only the first N rows (great for large files)
df_small = pd.read_csv("World_Dev_Indicators.csv", nrows=100)
print(f"Loaded {len(df_small)} rows")
df_small.head(3)
```

```python
# usecols — load only specific columns (saves memory)
cols_to_load = ["Year", "Country", "GDP (current US$)", "Region"]
df_subset = pd.read_csv("World_Dev_Indicators.csv", usecols=cols_to_load)
print(f"Columns loaded: {df_subset.columns.tolist()}")
df_subset.head(3)
```

```python
# dtype — specify data types at load time
df_typed = pd.read_csv(
    "World_Dev_Indicators.csv",
    dtype={"Year": str, "Country Code": str}   # force Year to string
)
print("Year dtype:", df_typed["Year"].dtype)
df_typed[["Year","Country","Country Code"]].head(3)
```

```python
# na_values — treat custom strings as NaN (missing)
df_na = pd.read_csv(
    "World_Dev_Indicators.csv",
    na_values=["N/A", "--", "missing"]  # extra patterns to treat as NaN
)
print("Loaded with custom NA values")
df_na.head(3)
```

## 4. Exporting Data

After analysis, we often need to save results. Pandas makes it easy.

### 4.1 Export to CSV

```python
# Save to CSV — index=False avoids writing the row numbers as a column
df.to_csv("output_full.csv", index=False)
print("Saved: output_full.csv")

# Save only specific columns
df[["Year","Country","GDP (current US$)"]].to_csv("output_gdp_only.csv", index=False)
print("Saved: output_gdp_only.csv")
```

### 4.2 Export to Excel

```python
# Export to Excel (requires: pip install openpyxl)
df.to_excel("output_full.xlsx", index=False, sheet_name="WDI Data")
print("Saved: output_full.xlsx")
```

```python
# Export multiple DataFrames to different sheets in one Excel file
with pd.ExcelWriter("output_multisheet.xlsx") as writer:
    df[df["Region"].str.strip() == "South Asia"].to_excel(
        writer, sheet_name="South Asia", index=False)
    df[df["Region"].str.strip() == "North America"].to_excel(
        writer, sheet_name="North America", index=False)
    df[df["Income Group"] == "High income"].to_excel(
        writer, sheet_name="High Income", index=False)
print("Saved: output_multisheet.xlsx with 3 sheets")
```

### 4.3 Export to JSON

```python
# Export to JSON (one record per row)
df.head(20).to_json("output_sample.json", orient="records", indent=2)
print("Saved: output_sample.json")

# Read it back to verify
import json
with open("output_sample.json") as f:
    data = json.load(f)
print(f"JSON contains {len(data)} records")
print("Keys in first record:", list(data[0].keys()))
```

## 5. Reading Back Exported Files

Good practice: always verify your exports by reading them back.

```python
# Re-load the CSV we just exported and check it matches
df_check = pd.read_csv("output_full.csv")
print("Re-loaded CSV shape:", df_check.shape)
print("Columns match original:", list(df_check.columns) == list(df.columns))
df_check.head(3)
```

## 6. Summary Cheat Sheet

| Task | Code |
|------|------|
| Load CSV | `pd.read_csv('file.csv')` |
| Load first N rows | `pd.read_csv('file.csv', nrows=N)` |
| Load specific columns | `pd.read_csv('file.csv', usecols=['col1','col2'])` |
| Save to CSV | `df.to_csv('out.csv', index=False)` |
| Save to Excel | `df.to_excel('out.xlsx', index=False)` |
| Save to JSON | `df.to_json('out.json', orient='records', indent=2)` |
