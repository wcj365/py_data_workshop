## 1. Load the Data

We begin every session the same way — import pandas and load our dataset.

```python
import pandas as pd

df = pd.read_csv("World_Dev_Indicators.csv")
print(f"Loaded {df.shape[0]:,} rows and {df.shape[1]} columns")
df.head()
```

## 2. Shape and Size

The very first questions to answer about any dataset:
- How many **rows** (observations)?
- How many **columns** (variables)?

```python
# .shape returns a tuple (rows, columns)
print("Shape:", df.shape)
print(f"  Rows:    {df.shape[0]:,}")
print(f"  Columns: {df.shape[1]}")

# Total number of data cells
print(f"  Total cells: {df.size:,}")
```

## 3. Data Types with `.dtypes` and `.info()`

Every column in a DataFrame has a **dtype** — the kind of data it stores.
Knowing dtypes helps you understand what operations are available and spot encoding errors.

| Dtype | Meaning |
|-------|---------|
| `int64` | Whole numbers |
| `float64` | Decimal numbers |
| `object` | Text / strings |
| `bool` | True/False |
| `datetime64` | Dates and times |

```python
# See dtypes for each column
print(df.dtypes)
```

```python
# .info() gives a compact summary: shape, dtypes, AND non-null counts
df.info()
```

### 3.1 What does the dtype tell us?

Notice:
- `Year` is `int64` — that is correct, years are whole numbers
- GDP columns are `float64` — correct, monetary values have decimals
- `Country`, `Region`, `Income Group` are `object` (= string) — correct
- `Population, total` is `int64` — whole number, correct

If a numeric column shows up as `object`, that is a red flag — likely due to
commas in numbers or stray text characters. We will fix those in Session 4.

## 4. Statistical Summary with `.describe()`

`.describe()` computes count, mean, min, max, and quartiles for every numeric column.
This is one of the fastest ways to understand your data at a glance.

```python
# Describe all numeric columns
df.describe()
```

```python
# Describe only specific columns
df[["GDP (current US$)", "GDP per capita (current US$)", "Life expectancy at birth, total (years)"]].describe()
```

```python
# describe() on text columns — shows unique, top, and frequency
df[["Country","Region","Income Group","Lending Type"]].describe()
```

```python
# Transpose for easier reading when there are many columns
df.describe().T
```

## 5. Unique Values and Frequencies

For categorical columns (like Region, Income Group), we want to know
**how many unique values** exist and **how often each appears**.

```python
# .nunique() — count unique values per column
print("Unique values per column:")
print(df.nunique())
```

```python
# .unique() — list all unique values in a column
print("Unique Regions:")
for r in df["Region"].unique():
    print(" ", r)
```

```python
# .value_counts() — count how often each value appears
print("Observations per Income Group:")
print(df["Income Group"].value_counts())
print()
print("Observations per Region:")
print(df["Region"].value_counts())
```

```python
# value_counts with normalize=True → proportions (0 to 1)
print("Proportion by Income Group:")
print(df["Income Group"].value_counts(normalize=True).round(3))
```

## 6. Inspecting Specific Columns

Sometimes you want a closer look at a single column.

```python
# Year range — min and max
print(f"Years covered: {df['Year'].min()} to {df['Year'].max()}")
print(f"Number of distinct years: {df['Year'].nunique()}")
print("Years list:", sorted(df["Year"].unique()))
```

```python
# Countries in the dataset
print(f"Number of distinct countries: {df['Country'].nunique()}")
print("\nFirst 10 countries (alphabetical):")
print(sorted(df["Country"].unique())[:10])
```

```python
# Quick stats on a single numeric column
col = "Life expectancy at birth, total (years)"
print(f"Life Expectancy Stats:")
print(f"  Min:  {df[col].min():.1f} years")
print(f"  Max:  {df[col].max():.1f} years")
print(f"  Mean: {df[col].mean():.1f} years")
print(f"  Median: {df[col].median():.1f} years")
```

## 7. A Quick Data Profile

Let us write a simple function that produces a one-stop summary for any DataFrame.

```python
def quick_profile(df):
    """Print a quick profile of a DataFrame."""
    print("=" * 55)
    print(f"DATASET PROFILE")
    print("=" * 55)
    print(f"Rows:    {df.shape[0]:,}")
    print(f"Columns: {df.shape[1]}")
    print(f"Memory:  {df.memory_usage(deep=True).sum() / 1e6:.2f} MB")
    print()
    print("Column Summary:")
    print(f"{'Column':<45} {'Dtype':<12} {'Nulls':<8} {'Unique'}")
    print("-" * 75)
    for col in df.columns:
        nulls = df[col].isnull().sum()
        unique = df[col].nunique()
        print(f"{col:<45} {str(df[col].dtype):<12} {nulls:<8} {unique}")
    print("=" * 55)

quick_profile(df)
```
